```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
le=LabelEncoder()
data['financial_institution']=le.fit_transform(data['financial_institution'])
data['loan_purpose']=le.fit_transform(data['loan_purpose'])
data['source']=le.fit_transform(data['source'])
```

```python
train = data.loc[data['type']=='train']
test = data.loc[data['type']=='test']
```

```python
data.columns
```

```python
data.info()
```

```python
def feature_scaling(df):
    df -= df.min()
    df /= df.max()
    return df
data[['borrower_credit_score','co-borrower_credit_score','debt_to_income_ratio','loan_to_value','unpaid_principal_bal']]=feature_scaling(data[['borrower_credit_score','co-borrower_credit_score','debt_to_income_ratio','loan_to_value','unpaid_principal_bal']])
```

```python
train=train.drop(['type','first_payment_date','loan_id','origination_date'],axis=1)
test=test.drop(['type','first_payment_date','loan_id','origination_date','m13'],axis=1)
```

```python
X_train=train.drop('m13',axis=1)
Y_train=train['m13']
X_test=test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import BaggingClassifier

ada=ExtraTreesClassifier()
ada.fit(X_train, Y_train)
Y_pred2 = ada.predict(X_test)

```

```python
ada.score(X_train,Y_train)
```
